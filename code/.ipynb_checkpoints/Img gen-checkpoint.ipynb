```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import tifffile as tiff
import cv2
import os
from tqdm.notebook import tqdm
import rasterio
from rasterio.windows import Window
from torch.utils.data import Dataset
import gc
```

```python
MASKS = '../data/HuBMAP/train.csv'
DATA = '../data/HuBMAP/train/'
```

```python
#functions to convert encoding to mask and mask to encoding
def enc2mask(encs, shape):
    img = np.zeros(shape[0]*shape[1], dtype=np.uint8)
    for m,enc in enumerate(encs):
        if isinstance(enc, float) and np.isnan(enc): continue
        s = enc.split()
        for i in range(len(s)//2):
            start = int(s[2*i]) - 1
            length = int(s[2*i+1])
            img[start:start+length] = 1 + m
    return img.reshape(shape).T

def mask2enc(mask, n=1):
    pixels = mask.T.flatten()
    encs = []
    for i in range(1,n+1):
        p = (pixels == i).astype(np.int8)
        if p.sum() == 0: encs.append(np.nan)
        else:
            p = np.concatenate([[0], p, [0]])
            runs = np.where(p[1:] != p[:-1])[0] + 1
            runs[1::2] -= runs[::2]
            encs.append(' '.join(str(x) for x in runs))
    return encs

df_masks = pd.read_csv(MASKS).set_index('id')
df_masks
```

```python
# 데이터를 섞어줍니다.
df_masks = df_masks.sample(frac=1, random_state=42)

# 데이터를 train, valid, test로 분할합니다.
train_df_masks = df_masks.iloc[:12]
valid_df_masks = df_masks.iloc[12:14]
test_df_masks = df_masks.iloc[[14]]
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=5, shuffle=True, random_state=42)

folds_data = {}
for fold, (train_idx, test_idx) in enumerate(kf.split(df_masks)):
    folds_data[fold] = {
        'train': df_masks.iloc[train_idx],
        'test': df_masks.iloc[test_idx]
    }
```

```python
#one of the new images cannot be loaded into 16GB RAM
#use rasterio to load image part by part
#using a dataset similar to my submission kernel

class HuBMAPDataset(Dataset):
    def __init__(self, idx, sz, reduce, s_th, p_th, encs=None):
        self.data = rasterio.open(os.path.join(DATA,idx+'.tiff'),num_threads='all_cpus')
        # some images have issues with their format 
        # and must be saved correctly before reading with rasterio
        if self.data.count != 3:
            subdatasets = self.data.subdatasets
            self.layers = []
            if len(subdatasets) > 0:
                for i, subdataset in enumerate(subdatasets, 0):
                    self.layers.append(rasterio.open(subdataset))
        self.shape = self.data.shape
        self.reduce = reduce
        self.sz = reduce*sz
        self.pad0 = (self.sz - self.shape[0]%self.sz)%self.sz
        self.pad1 = (self.sz - self.shape[1]%self.sz)%self.sz
        self.n0max = (self.shape[0] + self.pad0)//self.sz
        self.n1max = (self.shape[1] + self.pad1)//self.sz
        self.mask = enc2mask(encs,(self.shape[1],self.shape[0])) if encs is not None else None
        self.s_th = s_th
        self.p_th = p_th
        
    def __len__(self):
        return self.n0max*self.n1max
    
    def __getitem__(self, idx):
        # the code below may be a little bit difficult to understand,
        # but the thing it does is mapping the original image to
        # tiles created with adding padding (like in the previous version of the kernel)
        # then the tiles are loaded with rasterio
        # n0,n1 - are the x and y index of the tile (idx = n0*self.n1max + n1)
        n0,n1 = idx//self.n1max, idx%self.n1max
        # x0,y0 - are the coordinates of the lower left corner of the tile in the image
        # negative numbers correspond to padding (which must not be loaded)
        x0,y0 = -self.pad0//2 + n0*self.sz, -self.pad1//2 + n1*self.sz

        # make sure that the region to read is within the image
        p00,p01 = max(0,x0), min(x0+self.sz,self.shape[0])
        p10,p11 = max(0,y0), min(y0+self.sz,self.shape[1])
        img = np.zeros((self.sz,self.sz,3),np.uint8)
        mask = np.zeros((self.sz,self.sz),np.uint8)
        # mapping the loade region to the tile
        if self.data.count == 3:
            img[(p00-x0):(p01-x0),(p10-y0):(p11-y0)] = np.moveaxis(self.data.read([1,2,3],
                window=Window.from_slices((p00,p01),(p10,p11))), 0, -1)
        else:
            for i,layer in enumerate(self.layers):
                img[(p00-x0):(p01-x0),(p10-y0):(p11-y0),i] =\
                  layer.read(1,window=Window.from_slices((p00,p01),(p10,p11)))
        if self.mask is not None: mask[(p00-x0):(p01-x0),(p10-y0):(p11-y0)] = self.mask[p00:p01,p10:p11]
        
        if self.reduce != 1:
            img = cv2.resize(img,(self.sz//self.reduce, self.sz//self.reduce),
                             interpolation = cv2.INTER_AREA)
            mask = cv2.resize(mask,(self.sz//self.reduce, self.sz//self.reduce),
                             interpolation = cv2.INTER_NEAREST)
        #check for empty imges
        hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
        h,s,v = cv2.split(hsv)
        #return -1 for empty images
        return img, mask, (-1 if (s>self.s_th).sum() <= self.p_th or img.sum() <= self.p_th else idx)
```

```python
def generate_images(sz=int, reduce=int, fold=int):
    origin_scale = 40
    scale = int(origin_scale/reduce)

    #print("x"+str(scale), "processing ...")
    
    s_th = 40  #saturation blancking threshold
    p_th = 1000*(sz//256)**2 #threshold for the minimum number of pixels

    OUT_TRAIN = f'../data/x{scale}/Fold{fold}/TRAIN/'

    x_tot,x2_tot = [],[]

    for index, encs in tqdm(folds_data[fold]['train'].iterrows(), total=len(folds_data[fold]['train'])):
        ds = HuBMAPDataset(index, sz=sz, reduce=reduce, s_th=s_th, p_th=p_th, encs=encs)
        for i in range(len(ds)):
            im, m, idx = ds[i]
            if idx < 0: continue

            x_tot.append((im/255.0).reshape(-1, 3).mean(0))
            x2_tot.append(((im/255.0)**2).reshape(-1, 3).mean(0))

            OUT_IMG = os.path.join(OUT_TRAIN, 'img/')
            OUT_MASK = os.path.join(OUT_TRAIN, 'mask/')

            os.makedirs(OUT_IMG, exist_ok=True)
            os.makedirs(OUT_MASK, exist_ok=True)

            img_path = os.path.join(OUT_IMG, f'{index}_{idx:04d}.png')
            mask_path = os.path.join(OUT_MASK, f'{index}_{idx:04d}.png')

            cv2.imwrite(img_path, cv2.cvtColor(im, cv2.COLOR_RGB2BGR))
            cv2.imwrite(mask_path, m)
        
    img_avr =  np.array(x_tot).mean(0)
    img_std =  np.sqrt(np.array(x2_tot).mean(0) - img_avr**2)
    print('mean:',img_avr, ', std:', img_std)
    
    OUT_TEST = f'../data/x{scale}/Fold{fold}/TEST/'

    x_tot,x2_tot = [],[]

    for index, encs in tqdm(folds_data[fold]['test'].iterrows(), total=len(folds_data[fold]['test'])):
        ds = HuBMAPDataset(index, sz=sz, reduce=reduce, s_th=s_th, p_th=p_th, encs=encs)
        for i in range(len(ds)):
            im, m, idx = ds[i]
            if idx < 0: continue

            x_tot.append((im/255.0).reshape(-1, 3).mean(0))
            x2_tot.append(((im/255.0)**2).reshape(-1, 3).mean(0))

            OUT_IMG = os.path.join(OUT_TEST, 'img/')
            OUT_MASK = os.path.join(OUT_TEST, 'mask/')

            os.makedirs(OUT_IMG, exist_ok=True)
            os.makedirs(OUT_MASK, exist_ok=True)

            img_path = os.path.join(OUT_IMG, f'{index}_{idx:04d}.png')
            mask_path = os.path.join(OUT_MASK, f'{index}_{idx:04d}.png')

            cv2.imwrite(img_path, cv2.cvtColor(im, cv2.COLOR_RGB2BGR))
            cv2.imwrite(mask_path, m)
        
    img_avr =  np.array(x_tot).mean(0)
    img_std =  np.sqrt(np.array(x2_tot).mean(0) - img_avr**2)
    print('mean:',img_avr, ', std:', img_std)
```

```python
generate_images(sz=256, reduce=20, fold=0)
generate_images(sz=256, reduce=13, fold=0)
generate_images(sz=256, reduce=10, fold=0)
```
